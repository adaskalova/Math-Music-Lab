## Part 1: Physics of Sound

### 1.1 What is Sound?

Sound is a **mechanical wave** that propagates through a medium (air, water, solid materials). In the context of music, we're interested in **periodic waves** — waves that repeat regularly over time.

The simplest periodic wave is a **sine wave** (sinusoid), described by the equation:

$$y(t) = A \cdot \sin(2\pi f t)$$

Where:
- $y(t)$ — displacement at time $t$
- $A$ — **amplitude** (determines loudness)
- $f$ — **frequency** in Hz (determines pitch)
- $t$ — time in seconds

### 1.2 Key Parameters

| Parameter | Symbol | Unit | Musical Effect |
|-----------|--------|------|----------------|
| Frequency | $f$ | Hz | Pitch (higher frequency = higher note) |
| Amplitude | $A$ | - | Loudness (higher amplitude = louder) |
| Period | $T = 1/f$ | seconds | Time for one complete cycle |
| Wavelength | $\lambda = v/f$ | meters | Distance between wave peaks |

**Example:** The note A4 (standard tuning reference) has a frequency of **440 Hz**, meaning the wave completes 440 cycles per second.

---

### 1.2.1 Key Properties of a Sound Wave

Beyond frequency and amplitude, a sound wave has several important properties:

**Period ($T$)** — the time for one complete cycle of the wave:
$$T = \frac{1}{f}$$

For A4 (440 Hz): $T = \frac{1}{440} \approx 0.00227$ seconds (about 2.3 milliseconds).

**Phase ($\varphi$)** — the initial offset of the wave at time $t = 0$. The full equation becomes:
$$y(t) = A \cdot \sin(2\pi f t + \varphi)$$

Phase determines *where* in its cycle the wave starts. Two waves with the same frequency but different phases can reinforce or cancel each other.

**Wavelength ($\lambda$)** — the physical distance of one complete cycle:
$$\lambda = \frac{v}{f}$$

where $v$ is the speed of sound in the medium.

| Medium | Speed of sound | Wavelength of A4 (440 Hz) |
|--------|---------------|---------------------------|
| Air (20°C) | 343 m/s | 0.78 m |
| Water | 1,480 m/s | 3.36 m |
| Steel | 5,960 m/s | 13.55 m |

> **Key insight:** The same note has a different wavelength depending on the medium, but the frequency stays the same.

### 1.3 Visualizing a Pure Tone

Let's generate and visualize the note **A4 (440 Hz)** — the standard tuning reference used by orchestras worldwide.

```python
# Parameters for generating A4 musical note
frequency = 440   # Standard pitch for A4 note in Hz
amplitude = 1.0   # Maximum amplitude (0.0 to 1.0)
duration = 0.01   # Duration in seconds (10ms for visualization)
sample_rate = 44100  # Standard CD-quality audio sample rate (samples/second)

# Generate time axis array with evenly spaced values
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

# Generate sine wave: y(t) = A * sin(2πft)
y = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the generated sine wave
plt.figure(figsize=(12, 4))
plt.plot(t * 1000, y, color='#2196F3', linewidth=2)
plt.title('Pure Tone: A4 = 440 Hz', fontsize=14, fontweight='bold')
plt.xlabel('Time (ms)')   # X-axis in milliseconds for better readability
plt.ylabel('Amplitude')   # Y-axis represents wave amplitude
plt.axhline(y=0, color='black', linewidth=0.5, linestyle='--', alpha=0.3)   # Add reference line at y=0
plt.grid(True, alpha=0.3)   # Enable grid with low opacity
plt.tight_layout()          # Automatically adjust subplot parameters
plt.show()

# Calculate and print the wave period
period = 1 / frequency   # Period = 1 / frequency (in seconds)
print(f"Period of A4: T = 1/{frequency} = {period*1000:.4f} ms")   # Convert to milliseconds
```

#### Why 440 Hz? A Brief History

The choice of A4 = 440 Hz as the standard tuning pitch is relatively modern:

- In the **Baroque era** (1600–1750), A4 ranged from 392 to 466 Hz depending on the region
- In **1858**, the French government standardized A4 = 435 Hz
- In **1939**, an international conference in London proposed A4 = 440 Hz
- In **1955**, the International Organization for Standardization adopted it as **ISO 16**

Some orchestras today still tune slightly higher (e.g., Berlin Philharmonic uses A4 = 443 Hz) for a brighter sound.

### 1.4 Effect of Frequency on Pitch

Different frequencies produce different pitches. Let's compare three notes:
- **C4** (Middle C) = 261.63 Hz
- **A4** = 440 Hz
- **A5** (one octave above A4) = 880 Hz

Notice that **A5 has exactly double the frequency of A4** — this is the mathematical definition of an **octave**.

```python
# Compare different musical frequencies
frequencies = [261.63, 440, 880]             # C4 (Middle C), A4, A5 frequencies
colors = ['#4CAF50', '#2196F3', '#F44336']   # Distinct colors for each wave
labels = ['C4 (261.63 Hz)', 'A4 (440 Hz)', 'A5 (880 Hz)']   # Descriptive labels

plt.figure(figsize=(14, 5))
# Generate and plot each frequency
for freq, color, label in zip(frequencies, colors, labels):
    y_temp = np.sin(2 * np.pi * freq * t)
    plt.plot(t * 1000, y_temp, color=color, linewidth=1.5, label=label, alpha=0.8)

plt.title('Comparison of Different Frequencies', fontsize=14, fontweight='bold')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.legend(fontsize=11, loc='upper right')
plt.axhline(y=0, color='black', linewidth=0.5, linestyle='--', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("Observation: Higher frequency → more cycles in the same time period → higher pitch")
```

#### Logarithmic Perception of Pitch

The human ear perceives pitch **logarithmically**, not linearly. This means:

- The difference between 100 Hz and 200 Hz sounds like **one octave**
- The difference between 1000 Hz and 1100 Hz sounds like a **small step**

Even though both differences are 100 Hz, the first one sounds much larger. Our brain hears **ratios**, not absolute differences:

$$\text{Perceived pitch} \propto \log_2(f)$$

This is why musical scales are built on **multiplicative ratios** (×2 for octave, ×3/2 for fifth) rather than additive steps.

#### Decibels — Measuring Loudness

Loudness is also perceived logarithmically. The **decibel** scale measures sound intensity relative to a reference level:

$$L = 20 \cdot \log_{10}\left(\frac{A}{A_0}\right) \text{ dB}$$

where $A_0$ is the reference amplitude (threshold of hearing).

| Sound | Level (dB) |
|-------|------------|
| Threshold of hearing | 0 dB |
| Whisper | 20 dB |
| Normal conversation | 60 dB |
| Rock concert | 110 dB |
| Threshold of pain | 130 dB |

> **Key insight:** Every +6 dB means the amplitude doubles. Every +10 dB sounds roughly twice as loud to the human ear.

### 1.5 Effect of Amplitude on Loudness

Amplitude controls the **loudness** of a sound, not its pitch. Let's visualize the same frequency (440 Hz) with different amplitudes.

```python
# Compare different amplitudes
amplitudes = [0.3, 0.7, 1.0]                     # Low, medium, and high amplitude
colors_amp = ['#FF9800', '#9C27B0', '#2196F3']   # Color coding for amplitudes

plt.figure(figsize=(14, 5))
# Generate and plot each amplitude
for amp, color in zip(amplitudes, colors_amp):
    y_temp = amp * np.sin(2 * np.pi * 440 * t)   # Fixed frequency, varying amplitude
    plt.plot(t * 1000, y_temp, color=color, linewidth=1.5, label=f'Amplitude = {amp}', alpha=0.8)

plt.title('Effect of Amplitude on Sound (440 Hz)', fontsize=14, fontweight='bold')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.legend(fontsize=11)
plt.axhline(y=0, color='black', linewidth=0.5, linestyle='--', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("Observation: Higher amplitude → louder sound (same pitch)")
```

#### Amplitude vs. Intensity

It is important to distinguish between **amplitude** and **intensity**:

- **Amplitude** ($A$) — the maximum displacement of the wave
- **Intensity** ($I$) — the power carried by the wave per unit area

They are related by:
$$I \propto A^2$$

This means **doubling the amplitude quadruples the intensity**. This is why small changes in amplitude can have a big effect on perceived loudness.

The intensity of sound is measured in watts per square meter ($\text{W/m}^2$):
- Threshold of hearing: $I_0 = 10^{-12}$ W/m²
- Threshold of pain: $I \approx 1$ W/m²

That's a range of **12 orders of magnitude** — which is exactly why we use the logarithmic decibel scale.

### 1.6 Beating Phenomenon

When two frequencies that are **very close** to each other are played simultaneously, they create a pulsating effect called **beating**.

The beat frequency is:
$$f_{\text{beat}} = |f_1 - f_2|$$

This phenomenon is used by musicians to **tune instruments** — when the beating disappears, the instruments are in tune.

#### The Envelope Formula

When two tones with frequencies $f_1$ and $f_2$ are played together (with equal amplitude $A$), the resulting wave can be rewritten using a trigonometric identity:

$$y(t) = 2A \cos\left(\pi \Delta f \cdot t\right) \cdot \sin\left(2\pi \bar{f} \cdot t\right)$$

where:
- $\Delta f = |f_1 - f_2|$ — the frequency difference (controls the **beating speed**)
- $\bar{f} = \frac{f_1 + f_2}{2}$ — the average frequency (the pitch we actually hear)

The $\cos(\pi \Delta f \cdot t)$ term is the **envelope** — it modulates the amplitude, creating the pulsating effect.

#### Practical Example: Tuning a Guitar

When tuning a guitar string to match a reference pitch:

1. Play both tones simultaneously
2. If you hear **fast beating** → the string is far from the target
3. Adjust the string tension → beating slows down
4. When the **beating disappears** → the frequencies match perfectly ($\Delta f = 0$)

For example, if the reference is 440 Hz and the string plays 444 Hz:
- Beat frequency: $|440 - 444| = 4$ Hz (4 pulses per second)
- You tighten/loosen until the pulses stop

```python
# Demonstrate beating
f1, f2 = 440, 446  # Two close frequencies that will produce audible beating
t_long = np.linspace(0, 1.0, int(sample_rate * 1.0), endpoint=False)   # Longer time array (1 second)
# Generate individual waves
y1 = np.sin(2 * np.pi * f1 * t_long)
y2 = np.sin(2 * np.pi * f2 * t_long)
y_sum = y1 + y2   # Combined wave

# Plot only first 500ms for clarity
plot_duration = 0.5
plot_samples = int(sample_rate * plot_duration)

plt.figure(figsize=(14, 5))
plt.plot(t_long[:plot_samples] * 1000, y_sum[:plot_samples], color='#E91E63', linewidth=0.8)
plt.title(f'Beating: {f1} Hz + {f2} Hz (Beat frequency = {abs(f1-f2)} Hz)',
         fontsize=14, fontweight='bold')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.axhline(y=0, color='black', linewidth=0.5, linestyle='--', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
# Print beat frequency calculation
print(f"Beat frequency: |{f1} - {f2}| = {abs(f1-f2)} Hz")
print(f"You would hear {abs(f1-f2)} pulses per second")
```

### 1.7 Generating Audio Files

Now let's create actual sound files that we can listen to. We'll define two utility functions:
1. `generate_tone()` — creates a sine wave signal
2. `save_wav()` — saves the signal as a .wav file

```python
def generate_tone(freq, duration=1.0, sample_rate=44100, amplitude=0.5):
    """
    Generate a pure sine wave tone.

    Parameters:
    -----------
    freq : float
        Frequency in Hz
    duration : float
        Duration in seconds
    sample_rate : int
        Samples per second (44100 is CD quality)
    amplitude : float
        Amplitude (0.0 to 1.0)

    Returns:
    --------
    signal : numpy array
        The generated audio signal
    """
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    signal = amplitude * np.sin(2 * np.pi * freq * t)
    return signal


def save_wav(filename, signal, sample_rate=44100):
    """
    Save audio signal as a .wav file.

    Parameters:
    -----------
    filename : str
        Output filename (should end with .wav)
    signal : numpy array
        Audio signal (values between -1 and 1)
    sample_rate : int
        Samples per second
    """
    # Convert float signal to 16-bit integers (-32768 to 32767)
    signal_int = np.int16(signal * 32767)
    # Create and configure WAV file
    with wave.open(filename, 'w') as wav_file:
        wav_file.setnchannels(1)  # Mono audio
        wav_file.setsampwidth(2)  # 2 bytes per sample (16-bit)
        wav_file.setframerate(sample_rate)   # Set sample rate
        # Write each sample to file
        for sample in signal_int:
            # Pack as short integer
            wav_file.writeframes(struct.pack('h', sample))

print("✅ Audio utility functions defined")
```

```python
# Generate three musical tones as WAV files
tones = {
    # Middle C
    'tone_C4_261hz.wav': 261.63,
    # Standard tuning A
    'tone_A4_440hz.wav': 440.0,
    # One octave above A4
    'tone_A5_880hz.wav': 880.0
}
# Generate and save each tone
for filename, freq in tones.items():
    signal = generate_tone(freq, duration=1.0)
    save_wav(filename, signal)
    print(f"✅ Generated: {filename} ({freq} Hz)")

print("\n🎵 Audio files created! You can download and listen to them.")
```

### 1.8 Overtones and Timbre

Why do a piano and a violin sound different when playing the same note (e.g., A4 = 440 Hz)?

The answer is **overtones** (also called **harmonics**). Real instruments don't produce pure sine waves. Instead, they produce a **fundamental frequency** plus a series of higher frequencies:

$$y(t) = A_1 \sin(2\pi f t) + A_2 \sin(2\pi \cdot 2f \cdot t) + A_3 \sin(2\pi \cdot 3f \cdot t) + \ldots$$

| Harmonic | Frequency | Musical interval |
|----------|-----------|------------------|
| 1st (fundamental) | $f$ = 440 Hz | — |
| 2nd | $2f$ = 880 Hz | Octave |
| 3rd | $3f$ = 1320 Hz | Octave + Fifth |
| 4th | $4f$ = 1760 Hz | 2 Octaves |
| 5th | $5f$ = 2200 Hz | 2 Octaves + Major Third |

**Timbre** is determined by the **relative amplitudes** ($A_1, A_2, A_3, \ldots$) of these harmonics:

- **Flute** — mostly fundamental, very few overtones → pure, clean sound
- **Violin** — rich overtones → warm, complex sound
- **Trumpet** — strong odd harmonics → bright, brassy sound

> **Key insight:** Timbre is what allows us to distinguish instruments. Mathematically, it is the **shape of the frequency spectrum**.

### 1.9 Human Hearing Range

The human ear can perceive frequencies roughly between **20 Hz and 20,000 Hz** (20 kHz).

| Range | Frequency | Examples |
|-------|-----------|----------|
| Infrasound | < 20 Hz | Earthquakes, wind turbines |
| Bass | 20 – 250 Hz | Bass guitar, thunder |
| Mid-range | 250 – 4,000 Hz | Human voice, most instruments |
| Treble | 4,000 – 20,000 Hz | Cymbals, bird songs |
| Ultrasound | > 20,000 Hz | Dog whistles, medical imaging |

**How hearing changes with age:**
- At birth: up to ~20,000 Hz
- Age 25: upper limit drops to ~17,000 Hz
- Age 50: upper limit drops to ~12,000 Hz
- Age 70+: upper limit may be below 8,000 Hz

This is called **presbycusis** (age-related hearing loss) and primarily affects high frequencies.

**Musical range:**
- A standard piano covers **27.5 Hz (A0) to 4,186 Hz (C8)**
- The human voice ranges from about **85 Hz (bass) to 1,100 Hz (soprano)**
- But overtones extend much higher, which is why we need the full hearing range to appreciate music

> **Key insight:** The musical range is only a fraction of our hearing range, but overtones and harmonics fill the rest, giving music its richness.
