## 1. What is Consonance?

**Consonance** is the harmonious and pleasant sound produced by two or more tones perceived as stable and resolved.

**Mathematical Basis:** Simple frequency ratios create consonance:
$$\text{Consonance} \propto \frac{1}{\text{ratio complexity}}$$

**Types of Consonance:**
1. **Perfect Consonance** - Octave (2:1), Perfect Fifth (3:2), Perfect Fourth (4:3)
2. **Imperfect Consonance** - Major/Minor Thirds and Sixths
3. **Dissonance** - Seconds, Sevenths, Tritone

| Interval       | Ratio  | Type                 |
|----------------|--------|----------------------|
| Octave         | 2:1    | Perfect Consonance   |
| Perfect Fifth  | 3:2    | Perfect Consonance   |
| Perfect Fourth | 4:3    | Perfect Consonance   |
| Major Third    | 5:4    | Imperfect Consonance |
| Tritone        | 45:32  | Dissonance           |
| Minor Second   | 16:15  | Dissonance           |

## 2. What is Beating?

**Beating** is a physical phenomenon where two sound waves with close frequencies interact, creating periodic changes in loudness.

**Formula:**
$$f_{\text{beat}} = |f_1 - f_2|$$

where $f_1$ and $f_2$ are the frequencies of the two tones.

| Interval      | Frequencies (Hz) | Beat (Hz) | Effect            |
|---------------|-----------------|-----------|-------------------|
| Unison        | 440 and 440     | 0         | No beating        |
| Minor Second  | 440 and 466     | 26        | Strong beating    |
| Perfect Fifth | 440 and 660     | 220       | Weak beating      |
| Octave        | 440 and 880     | 440       | No perceived beat |

## 3. Role of Beating in Consonance

- **Consonant intervals** - weak or no beating (simple ratios, small frequency difference)
- **Dissonant intervals** - strong beating (complex ratios, large frequency difference)

**Helmholtz Theory:** Consonance results from the **absence of beating** between tones and their harmonics.
- Perfect Fifth (3:2): harmonics align, minimal beating
- Tritone (45:32): harmonics clash, strong beating

## 4. Relationship Between Consonance and Beating

1. **Helmholtz (1863):** Consonance = absence of beating between fundamental tones and overtones
2. **Psychoacoustics:** The brain perceives sounds with simple ratios as *pleasant*
3. **Critical Bandwidth:** Beating is most disturbing when $f_{beat}$ falls within the ear's critical bandwidth (~20-200 Hz)

## 5. Defining Intervals

> Add blockquote

```python
import numpy as np
import matplotlib.pyplot as plt
import wave

# Just Intonation frequency ratios
intervals = {
    'Unison':          1/1,
    'Minor Second':    16/15,
    'Major Second':    9/8,
    'Minor Third':     6/5,
    'Major Third':     5/4,
    'Perfect Fourth':  4/3,
    'Tritone':         45/32,
    'Perfect Fifth':   3/2,
    'Minor Sixth':     8/5,
    'Major Sixth':     5/3,
    'Minor Seventh':   16/9,
    'Major Seventh':   15/8,
    'Octave':          2/1
}

BASE_FREQ = 440    # A4 in Hz
SAMPLE_RATE = 44100

def generate_interval(base_freq=440, ratio=1.0, duration=2.0):
    """Generate two tones forming a musical interval."""
    t = np.linspace(0, duration, int(SAMPLE_RATE * duration), endpoint=False)
    tone1 = 0.5 * np.sin(2 * np.pi * base_freq * t)
    tone2 = 0.5 * np.sin(2 * np.pi * base_freq * ratio * t)
    return t, tone1, tone2, tone1 + tone2

# Print interval table
print(f"{'Interval':<20} {'Ratio':>10} {'Freq 1 (Hz)':>12} {'Freq 2 (Hz)':>12} {'Beat (Hz)':>10}")
print('-' * 68)
for name, ratio in intervals.items():
    f2 = BASE_FREQ * ratio
    beat = abs(BASE_FREQ - f2)
    print(f"{name:<20} {ratio:>10.4f} {BASE_FREQ:>12.2f} {f2:>12.2f} {beat:>10.2f}")
```

## 6. Visualizing Intervals

```python
# Visualize 6 key intervals
demo = ['Unison', 'Minor Second', 'Major Third', 'Perfect Fifth', 'Tritone', 'Octave']

fig, axes = plt.subplots(3, 2, figsize=(16, 12))
fig.suptitle('Part 3: Musical Intervals - Waveform Visualization', fontsize=16, fontweight='bold')

for i, name in enumerate(demo):
    ratio = intervals[name]
    t, tone1, tone2, combined = generate_interval(BASE_FREQ, ratio, duration=0.05)
    ax = axes[i // 2][i % 2]
    ax.plot(t[:500], tone1[:500], color='#3498db', alpha=0.7, label=f'Tone 1: {BASE_FREQ} Hz')
    ax.plot(t[:500], tone2[:500], color='#e74c3c', alpha=0.7, label=f'Tone 2: {BASE_FREQ*ratio:.1f} Hz')
    ax.plot(t[:500], combined[:500], color='#2c3e50', lw=1.5, label='Combined')
    ax.set_title(f'{name}  (ratio: {ratio:.4f})', fontsize=12)
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Amplitude')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 7. Consonance Spectrum

```python
consonance_levels = {
    'Unison':         0.99,
    'Octave':         0.95,
    'Perfect Fifth':  0.90,
    'Perfect Fourth': 0.85,
    'Major Sixth':    0.80,
    'Major Third':    0.75,
    'Minor Sixth':    0.70,
    'Minor Third':    0.65,
    'Major Second':   0.50,
    'Tritone':        0.35,
    'Minor Second':   0.30,
    'Major Seventh':  0.25,
    'Minor Seventh':  0.20
}

sorted_cons = sorted(consonance_levels.items(), key=lambda x: x[1], reverse=True)
names = [k for k, v in sorted_cons]
values = [v for k, v in sorted_cons]
colors = ['#2ecc71' if v > 0.7 else '#f1c40f' if v > 0.4 else '#e74c3c' for v in values]

plt.figure(figsize=(14, 7))
plt.bar(names, values, color=colors, alpha=0.85)
plt.title('Consonance Spectrum of Musical Intervals', fontsize=15, fontweight='bold')
plt.ylabel('Consonance Level (0-1)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1.1)
plt.axhline(0.7, color='green', linestyle='--', alpha=0.5, label='Consonance threshold')
plt.axhline(0.4, color='orange', linestyle='--', alpha=0.5, label='Dissonance threshold')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
