# Part 2: Musical Scales and Mathematics
## Understanding Tuning Systems and Frequency Relationships

In this section, explore the mathematical foundations of musical scales, comparing Pythagorean and Equal Temperament tuning systems.

### Historical Context of Musical Scales

The search for a "perfect" musical scale has occupied mathematicians and musicians for over 2,500 years:

| Era | System | Key Figure | Principle |
|-----|--------|------------|----------|
| ~500 BC | Pythagorean | Pythagoras | Stacking perfect fifths (3:2) |
| ~200 BC | Just Intonation | Ptolemy | Pure integer ratios (5:4, 6:5) |
| 1584 | Equal Temperament | Zhu Zaiyu (China) | $2^{n/12}$ — equal division of octave |
| 1636 | Equal Temperament | Marin Mersenne (Europe) | Independent rediscovery |
| 1722 | Practical adoption | J.S. Bach | *The Well-Tempered Clavier* demonstrated all 24 keys |

> **Key insight:** The mathematical problem is fundamental — it is **impossible** to have a scale that is simultaneously based on pure ratios AND allows free transposition between keys. Every tuning system is a compromise.

### Mathematical Foundations

**Pythagorean Tuning:**
$$f_n = f_0 \times \left(\frac{3}{2}\right)^k \times \left(\frac{1}{2}\right)^m$$

**Equal Temperament:**
$$f_n = f_0 \times 2^{n/12}$$

Where:
- $f_0$ = Base frequency (440Hz for A4)
- $k$ = Number of perfect fifths
- $m$ = Octave adjustment
- $n$ = Number of semitones from base

#### How Pythagorean Tuning is Built

Pythagorean tuning constructs all 12 notes by repeatedly applying the **perfect fifth** ratio (3:2), then folding back into a single octave:

1. Start with a base note (e.g., C)
2. Go up a fifth: $C \times \frac{3}{2} = G$
3. Go up another fifth: $G \times \frac{3}{2} = D'$ (then divide by 2 to stay in the same octave)
4. Continue 12 times to generate all notes

The sequence of fifths:
$$C \to G \to D \to A \to E \to B \to F\# \to C\# \to G\# \to D\# \to A\# \to F \to (C')$$

The problem: after 12 fifths, we expect to return to the starting note (7 octaves higher), but:
$$\left(\frac{3}{2}\right)^{12} = 129.746 \neq 2^7 = 128$$

This small difference ($\approx 1.36\%$) is the **Pythagorean Comma** — the fundamental reason why pure tuning cannot form a perfect closed cycle.

#### Why Equal Temperament Works

Equal temperament solves this by making every semitone **exactly** the same ratio:
$$r = 2^{1/12} \approx 1.05946$$

This means:
- Every semitone multiplies the frequency by the same factor
- 12 semitones = $r^{12} = (2^{1/12})^{12} = 2$ — a perfect octave, always
- The fifth becomes $r^7 = 2^{7/12} \approx 1.4983$ instead of the pure $1.5$

The trade-off: no interval is perfectly pure, but **all keys sound equally good** (or equally impure).

```python
import numpy as np # For numerical operations
import matplotlib.pyplot as plt # For data visualization

# Base frequency reference (A4 standard pitch)
A4 = 440  # Hz (International concert pitch standard)

# Pythagorean tuning ratios based on perfect fifths (3:2 ratios)
# Uses pure intervals derived from harmonic series
pythagorean_ratios = {
    'C': 1.0,       # Unison (reference note)
    'C#': 256/243,  # Limma (Pythagorean minor semitone)
    'D': 9/8,       # Major whole tone
    'D#': 32/27,    # Minor third approximation
    'E': 81/64,     # Major third approximation
    'F': 4/3,       # Perfect fourth
    'F#': 729/512,  # Apotome (Pythagorean major semitone)
    'G': 3/2,       # Perfect fifth (basis of the tuning)
    'G#': 128/81,   # Minor sixth approximation
    'A': 27/16,     # Major sixth approximation
    'A#': 16/9,     # Minor seventh
    'B': 243/128    # Major seventh
}

# Calculate Equal Temperament frequencies (modern standard)
# Divides octave into 12 equal logarithmic intervals
equal_temp = {note: A4 * 2**(i/12) for i, note in enumerate(pythagorean_ratios)}
```

### Frequency Comparison
Visual comparison between tuning systems:

```python
# Create figure for frequency comparison
plt.figure(figsize=(14, 7))

# Prepare data for plotting
notes = list(pythagorean_ratios.keys())   # Musical note names
pyth_freqs = [A4 * ratio for ratio in pythagorean_ratios.values()]   # Pythagorean frequencies
equal_freqs = [equal_temp[note] for note in notes]   # Equal temperament frequencies

# Create bar positions
x = np.arange(len(notes))
# Plot Pythagorean frequencies in blue
plt.bar(x - 0.2, pyth_freqs, width=0.4, label='Pythagorean', color='#3498db')
# Plot Equal Temperament frequencies in red
plt.bar(x + 0.2, equal_freqs, width=0.4, label='Equal Temperament', color='#e74c3c')
# Configure plot appearance
plt.xticks(x, notes)   # Set x-axis labels as note names
plt.title('Frequency Comparison: Pythagorean vs Equal Temperament')
plt.xlabel('Musical Notes')
plt.ylabel('Frequency (Hz)')
plt.legend()
plt.grid(True, alpha=0.3)   # Add grid with low opacity
plt.tight_layout()   # Adjust layout
plt.show()
```

### Cents Deviation Analysis
Measuring differences in cents (1 cent = 1/100 of a semitone):

#### Understanding Cents

The **cent** is a logarithmic unit for measuring musical intervals:

$$\text{cents} = 1200 \cdot \log_2\left(\frac{f_1}{f_2}\right)$$

Why use cents?
- One **semitone** in equal temperament = exactly **100 cents**
- One **octave** = exactly **1200 cents**
- The human ear can barely distinguish differences smaller than **5–10 cents**

| Difference | Perception |
|-----------|------------|
| < 5 cents | Imperceptible to most people |
| 5–15 cents | Noticeable by trained musicians |
| 15–25 cents | Clearly out of tune |
| > 25 cents | Obviously wrong |

The Pythagorean comma in cents:
$$1200 \cdot \log_2\left(\frac{(3/2)^{12}}{2^7}\right) = 1200 \cdot \log_2(1.01364) \approx 23.46 \text{ cents}$$

> This is clearly audible — which is why Pythagorean tuning sounds "off" in certain keys.

```python
# Function to calculate cents difference between frequencies
# Cents = 1200 × log₂(f₁/f₂) - standard unit for musical intervals
def cents_diff(f1, f2):
    """Calculate interval between two frequencies in cents"""
    return 1200 * np.log2(f1 / f2) if f1 != f2 else 0   # Avoid division by zero

# Calculate deviations between Pythagorean and Equal Temperament
deviations = [cents_diff(p, e) for p, e in zip(pyth_freqs, equal_freqs)]

# Create plot for deviations
plt.figure(figsize=(14, 6))
# Color coding: green for small deviations (<10 cents), red for larger ones
plt.bar(notes, deviations, color=['#27ae60' if abs(d) < 10 else '#e74c3c' for d in deviations])
plt.axhline(0, color='black', linewidth=0.8)   # Reference line at zero
plt.title('Cents Deviation: Pythagorean from Equal Temperament')
plt.xlabel('Notes')
plt.ylabel('Deviation (cents)')
plt.grid(True, alpha=0.3)
plt.show()
```

### Why 12 Notes?
The mathematical justification for the 12-tone system:

$$\left(\frac{3}{2}\right)^{12} \approx 129.746$$
$$2^7 = 128 \approx 129.746$$

The difference is just 1.3% - close enough to create a closed cycle:
$$\frac{129.746}{128} \approx 1\text{ octave}$$

This is known as the **Pythagorean Comma**.

### Modular Arithmetic in Music
Transposing notes using modulo 12:

```python
# Transposition function using chromatic circle
def transpose(note, semitones):
  """
        Transpose a note by specified number of semitones

        Parameters:
        note (str): Starting note (e.g., 'C')
        semitones (int): Number of semitones to transpose

        Returns:
        str: Transposed note name
    """
    note_index = notes.index(note)              # Get index of starting note
    new_index = (note_index + semitones) % 12   # Calculate new position modulo 12
    # Return transposed note
    return notes[new_index]

# Example: Transpose C by 7 semitones
print(f"Transposing C by 7 semitones: {transpose('C', 7)}")  # Perfect fifth (C → G)
print(f"Transposing E by 5 semitones: {transpose('E', 5)}")  # Perfect fourth (E → A)
```

### Conclusion
- Pythagorean tuning uses pure ratios but creates inconsistent intervals
- Equal temperament sacrifices purity for consistency across all keys
- The 12-tone system emerges from mathematical approximations of perfect fifths
- Modular arithmetic enables transposition and key changes
