# Part 2: Musical Scales and Mathematics
## Understanding Tuning Systems and Frequency Relationships

In this section, explore the mathematical foundations of musical scales, comparing Pythagorean and Equal Temperament tuning systems.

### Mathematical Foundations

**Pythagorean Tuning:**
$$f_n = f_0 \times \left(\frac{3}{2}\right)^k \times \left(\frac{1}{2}\right)^m$$

**Equal Temperament:**
$$f_n = f_0 \times 2^{n/12}$$

Where:
- $f_0$ = Base frequency (440Hz for A4)
- $k$ = Number of perfect fifths
- $m$ = Octave adjustment
- $n$ = Number of semitones from base

```python
import numpy as np
import matplotlib.pyplot as plt

# Base frequency
A4 = 440  # Hz

# Pythagorean tuning ratios
pythagorean_ratios = {
    'C': 1.0,
    'C#': 256/243,  # Limma
    'D': 9/8,
    'D#': 32/27,
    'E': 81/64,
    'F': 4/3,
    'F#': 729/512,  # Apotome
    'G': 3/2,
    'G#': 128/81,
    'A': 27/16,
    'A#': 16/9,
    'B': 243/128
}

# Equal temperament frequencies
equal_temp = {note: A4 * 2**(i/12) for i, note in enumerate(pythagorean_ratios)}
```

### Frequency Comparison
Visual comparison between tuning systems:

```python
plt.figure(figsize=(14, 7))

# Calculate frequencies
notes = list(pythagorean_ratios.keys())
pyth_freqs = [A4 * ratio for ratio in pythagorean_ratios.values()]
equal_freqs = [equal_temp[note] for note in notes]

# Plot
x = np.arange(len(notes))
plt.bar(x - 0.2, pyth_freqs, width=0.4, label='Pythagorean', color='#3498db')
plt.bar(x + 0.2, equal_freqs, width=0.4, label='Equal Temperament', color='#e74c3c')

plt.xticks(x, notes)
plt.title('Frequency Comparison: Pythagorean vs Equal Temperament')
plt.xlabel('Musical Notes')
plt.ylabel('Frequency (Hz)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Cents Deviation Analysis
Measuring differences in cents (1 cent = 1/100 of a semitone):

```python
def cents_diff(f1, f2):
    return 1200 * np.log2(f1 / f2) if f1 != f2 else 0

deviations = [cents_diff(p, e) for p, e in zip(pyth_freqs, equal_freqs)]

plt.figure(figsize=(14, 6))
plt.bar(notes, deviations, color=['#27ae60' if abs(d) < 10 else '#e74c3c' for d in deviations])
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Cents Deviation: Pythagorean from Equal Temperament')
plt.xlabel('Notes')
plt.ylabel('Deviation (cents)')
plt.grid(True, alpha=0.3)
plt.show()
```

### Why 12 Notes?
The mathematical justification for the 12-tone system:

$$\left(\frac{3}{2}\right)^{12} \approx 129.746$$
$$2^7 = 128 \approx 129.746$$

The difference is just 1.3% - close enough to create a closed cycle:
$$\frac{129.746}{128} \approx 1\text{ octave}$$

This is known as the **Pythagorean Comma**.

### Modular Arithmetic in Music
Transposing notes using modulo 12:

```python
def transpose(note, semitones):
    note_index = notes.index(note)
    new_index = (note_index + semitones) % 12
    return notes[new_index]

# Example: Transpose C by 7 semitones
print(f"Transposing C by 7 semitones: {transpose('C', 7)}")  # Should be G
print(f"Transposing E by 5 semitones: {transpose('E', 5)}")  # Should be A
```

### Conclusion
- Pythagorean tuning uses pure ratios but creates inconsistent intervals
- Equal temperament sacrifices purity for consistency across all keys
- The 12-tone system emerges from mathematical approximations of perfect fifths
- Modular arithmetic enables transposition and key changes
